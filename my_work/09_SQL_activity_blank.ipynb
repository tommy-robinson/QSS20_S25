```python
# your code here 1
cnx = mysql.connector.connect(user=creds['practice_database']['user'], 
                            password=creds['practice_database']['password'],
                            port=creds['practice_database']['port'],
                            database= creds['practice_database']['database'],
                            host = creds['practice_database']['host'])
cnx

new_col_q = """
SELECT *,
       CASE
         WHEN INCIDENT_CITY = 'Chicago' THEN 'YES'
         ELSE 'NO'
       END AS in_chicago_1
FROM   caseinit 
"""

new_col_d = pd.read_sql_query(new_col_q, cnx)
new_col_d
pd.crosstab(new_col_d['INCIDENT_CITY'], new_col_d['in_chicago_1'])
```

```python
# your code here 3: Use where to row filter to initiations in Chicago and use group by to find the count of 
# cases diverted and not diverted (is_in_diversion); pull the table with those counts
q = """
SELECT is_in_diversion, Count(*) as num_cases
FROM   caseinit 
WHERE INCIDENT_CITY = 'Chicago'
GROUP BY is_in_diversion
"""

new_col_d = pd.read_sql_query(q, cnx)
new_col_d
```

```python
# your code here 4: 4. Modify the query in step 3 to find the proportion of cases in chicago diverted (hint you made need to use case when
# in a subquery)
prop_div = """
SELECT 
    SUM(CASE WHEN is_in_diversion = 'True' THEN 1 ELSE 0 END) * 1.0 / COUNT(*) AS proportion_diverted
FROM (
    SELECT *,
           CASE 
               WHEN INCIDENT_CITY = 'Chicago' THEN 'YES' 
               ELSE 'NO' 
           END AS in_chicago2
    FROM caseinit
) AS tmp
WHERE in_chicago2 = 'YES';
"""

prop_div_d = pd.read_sql_query(prop_div, cnx)
prop_div_d
```

```python
# your code here 5: 6. Modify the query in step 4 to find the proportion of cases in chicago versus cases not in chicago sent to diversion 
query_propdiv_compare = """
SELECT 
    AVG(is_div_bool) as prop_div, 
    in_chicago_1
FROM  (
    SELECT 
        *,
        CASE 
            WHEN is_in_diversion = 'True' THEN 1
            ELSE 0
        END as is_div_bool,
        CASE 
            WHEN INCIDENT_CITY = 'Chicago' THEN 'Yes'
            ELSE 'No'
        END AS in_chicago_1
    FROM caseinit) AS d
WHERE INCIDENT_CITY <> ''
GROUP BY in_chicago_1
"""
prop_div_compare = pd.read_sql_query(query_propdiv_compare, cnx)
prop_div_compare
```
